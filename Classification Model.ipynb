```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("airlines_reviews.csv")
df
```

```python
label_encoder = LabelEncoder()

# Fit and transform the 'Class' column
df['Class'] = label_encoder.fit_transform(df['Class'])

# Verify the label encoding by displaying the first few rows
print(df.head())
```

```python
x = df.drop(["Title", "Name", "Review Date", "Airline", "Verified", "Reviews", "Type of Traveller", "Month Flown", "Route", "Recommended"], axis = 1)
```

```python
y = df["Recommended"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
```

```python
pred
```

```python
accuracy = accuracy_score(y_test, pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
recall = recall_score(y_test, pred ,pos_label='yes')
print(f"Recall: {recall:.2f}")
```

```python
precision = precision_score(y_test, pred, pos_label='yes')
print(f"Precision: {precision:.2f}")
```

```python
f1 = f1_score(y_test, pred, pos_label='yes')
print(f"F1 Score: {f1:.2f}")
```
